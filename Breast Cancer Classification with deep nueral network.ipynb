```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
```

```python
#loading dataset from sklearn
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
breast_cancer_dataset
```

```python
breast_cancer_dataset.feature_names
```

```python
#conver to dataframe
data=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
data
```

```python
data['target']=breast_cancer_dataset.target
data
```

```python
X=data.drop('target',axis=1)
y=data['target']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_data=scaler.fit_transform(X)
```

```python

xx=pd.DataFrame(X_data)
xx
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_data, y, test_size=0.2, random_state=2)
```

```python
import tensorflow as tf 
from tensorflow import keras
```

```python
model = keras.Sequential([
    
            keras.layers.Flatten(input_shape=(30,)),
            keras.layers.Dense(25 , activation = "relu"), 
            keras.layers.Dense(15 , activation = "relu") ,
            keras.layers.Dense(10 , activation = "relu") ,
            keras.layers.Dropout(0.2), 
            keras.layers.Dense(5 , activation = "relu") , 
            keras.layers.Dense(2 , activation = "sigmoid")
    
])
```

```python
model.compile(optimizer="adam" , metrics=['accuracy'] ,loss='sparse_categorical_crossentropy',)
```

```python
trained_model = model.fit(X_train, Y_train , epochs=25 )
```

```python
trained_model.history
```

```python
plt.plot(trained_model.history['accuracy'])
```

```python
plt.plot(trained_model.history['loss'])
```

```python
plt.plot(trained_model.history['accuracy'])
plt.plot(trained_model.history['loss'])
plt.title('model accuracy')
plt.xlabel('accuracy')
plt.ylabel('loss')
plt.legend(['training data', 'validation data'],loc='upper left')
plt.show()
```

```python
loss , acc = model.evaluate(X_test ,Y_test)
```
